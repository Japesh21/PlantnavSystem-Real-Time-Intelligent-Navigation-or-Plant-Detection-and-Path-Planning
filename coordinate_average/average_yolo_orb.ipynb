```python
import re

# Input lines (you can replace with actual inputs or dynamic input later)
yolo_text = "🪴 Potted plant coordinates: (177, 66), (378, 392)"
orb_text = "🔍 ORB matched coordinates (top 10): [(255, 186), (293, 182), (253, 288), (366, 53), (374, 181), (371, 285), (182, 178), (343, 229), (367, 289), (205, 277)]"

print("📥 Processing lines...")

# Extract YOLO top-left coordinate using regex
yolo_match = re.search(r"Potted plant coordinates: \((\d+), (\d+)\)", yolo_text)
if yolo_match:
    yolo_x, yolo_y = int(yolo_match.group(1)), int(yolo_match.group(2))
    print(f"✅ YOLO top-left: ({yolo_x}, {yolo_y})")
else:
    print("❌ No YOLO match found in:", yolo_text)

# Manually parse ORB matched coordinates
orb_coordinates = []
# Extract ORB coordinate part from the text
try:
    orb_match = re.search(r"\[\((\d+), (\d+)\)", orb_text)  # Regex to capture first coordinate pair
    if orb_match:
        orb_x, orb_y = int(orb_match.group(1)), int(orb_match.group(2))
        print(f"✅ ORB first match: ({orb_x}, {orb_y})")
    else:
        print("❌ No ORB match found in:", orb_text)
except Exception as e:
    print(f"❌ Error extracting ORB coordinates: {e}")

# If both found, average them
if yolo_match and orb_x is not None and orb_y is not None:
    avg_x = (yolo_x + orb_x) // 2
    avg_y = (yolo_y + orb_y) // 2
    print(f"🎯 Averaged coordinate: ({avg_x}, {avg_y})")
```
