```python
import os
import shutil

# Define your paths
image_folder = r"C:/Users/91939/OneDrive/Desktop/python/data/dataset/train"  # Path to train images
label_folder = r"C:/Users/91939/OneDrive/Desktop/python/data/yolo_labels/train1"  # Path to labels

# Iterate over all the subdirectories (like 'aloevera', 'banana', etc.) in the image folder
for subdir in os.listdir(image_folder):
    subdir_path = os.path.join(image_folder, subdir)
    if os.path.isdir(subdir_path):  # Check if it's a directory
        # Look for images and matching label files
        for image_file in os.listdir(subdir_path):
            if image_file.endswith(('.jpg', '.png', '.jpeg')):  # Ensure it's an image file
                image_name = os.path.splitext(image_file)[0]  # Get the image name without extension
                label_file = image_name + '.txt'  # The corresponding label file
                label_path = os.path.join(label_folder, label_file)  # Path to the label

                # If the label file exists, move it to the correct directory
                if os.path.exists(label_path):
                    dest_label_path = os.path.join(subdir_path, label_file)
                    shutil.move(label_path, dest_label_path)
                    print(f"Moved: {label_file} to {subdir_path}")
                else:
                    print(f"Label file not found for: {image_file}")
```

```python
import os
import shutil

# Base path
base_path = "C:/Users/91939/OneDrive/Desktop/python/data/dataset/"

# Classes to copy labels for
classes = ["aloevera", "banana"]

# Number of label files to copy per class
num_to_copy = 10

for cls in classes:
    train_labels_dir = os.path.join(base_path, "train", cls)
    val_labels_dir = os.path.join(base_path, "val", cls)

    # Make sure the val class directory exists
    os.makedirs(val_labels_dir, exist_ok=True)

    # Get .txt label files in the training folder
    label_files = [f for f in os.listdir(train_labels_dir) if f.endswith(".txt")]

    # Copy a few .txt files to the validation folder
    for label_file in label_files[:num_to_copy]:
        src = os.path.join(train_labels_dir, label_file)
        dst = os.path.join(val_labels_dir, label_file)
        shutil.copyfile(src, dst)
        print(f"Copied: {label_file} -> {val_labels_dir}")
```

```python
import os
import shutil
from glob import glob

# Define source and target
base = r"C:/Users/91939/OneDrive/Desktop/python/data/dataset"
classes = ["aloevera", "banana"]

for cls in classes:
    src = os.path.join(base, "train", cls)
    dst = os.path.join(base, "val", cls)
    os.makedirs(dst, exist_ok=True)

    # Get 10 .txt files
    txt_files = sorted(glob(os.path.join(src, "*.txt")))[:10]
    
    for txt_path in txt_files:
        # Copy .txt file
        shutil.copy(txt_path, dst)
        print(f"Copied label: {os.path.basename(txt_path)}")

        # Find corresponding .jpg image and copy
        img_path = txt_path.replace(".txt", ".jpg")
        if os.path.exists(img_path):
            shutil.copy(img_path, dst)
            print(f"Copied image: {os.path.basename(img_path)}")
        else:
            print(f"⚠️ Image not found for {txt_path}")
```

```python
import os
import shutil

# Define your base dataset directory
base_dir = r"C:\Users\91939\OneDrive\Desktop\python\data\dataset\train"
classes = ['aloevera', 'banana']  # add other class folders if needed

for cls in classes:
    img_dir = os.path.join(base_dir, cls)
    new_img_dir = os.path.join(base_dir, 'images', cls)
    new_label_dir = os.path.join(base_dir, 'labels', cls)

    os.makedirs(new_img_dir, exist_ok=True)
    os.makedirs(new_label_dir, exist_ok=True)

    for file in os.listdir(img_dir):
        file_path = os.path.join(img_dir, file)

        if file.lower().endswith(('.jpg', '.png', '.jpeg')):
            shutil.move(file_path, os.path.join(new_img_dir, file))
        elif file.lower().endswith('.txt'):
            shutil.move(file_path, os.path.join(new_label_dir, file))

print("Files reorganized successfully!")
```

```python
import os
import shutil

# Path to the class folder
class_folder = r"C:/Users/91939/OneDrive/Desktop/python/data/dataset/val/aloevera"
images_dir = os.path.join(os.path.dirname(class_folder), "images")
labels_dir = os.path.join(os.path.dirname(class_folder), "labels")

# Create target folders if they don't exist
os.makedirs(images_dir, exist_ok=True)
os.makedirs(labels_dir, exist_ok=True)

# Move files to correct folders
for filename in os.listdir(class_folder):
    src_path = os.path.join(class_folder, filename)
    if filename.endswith(".jpg") or filename.endswith(".png") or filename.endswith(".jpeg"):
        shutil.move(src_path, os.path.join(images_dir, filename))
    elif filename.endswith(".txt"):
        shutil.move(src_path, os.path.join(labels_dir, filename))

print("✅ Files moved to val/images/ and val/labels/")
```
